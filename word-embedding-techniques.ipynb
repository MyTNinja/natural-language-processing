### **Import libraries**

```python
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
warnings.filterwarnings('ignore')
nltk.download('punkt')
```

### **Embedding**

**Example 1**

```python
tokenized_sentences = [['hello','this','is','python','hello','this','is','java','hello','this','is','c']]
```

```python
model = Word2Vec(tokenized_sentences, min_count=1)
```

```python
print(model)
```

```python
print(tokenized_sentences)
```

```python
# vector representations
print(model.wv['hello'])
```

```python
#  words most similar to 'hello'
model.wv.most_similar('hello')
```

```python
#  words most similar
model.wv.most_similar('java')
```

**Example 2**

```python
sentences_X = [
    "Natural language processing is an important field of AI",
    "AI techniques include machine learning and deep learning",
    "Machine learning can analyze large datasets for insights",
    "Word2Vec is a popular technique for word embeddings"
]

tokenized_sentences_X = [word_tokenize(sentence.lower()) for sentence in sentences_X]
```

```python
model_X = Word2Vec(sentences=tokenized_sentences_X, vector_size=100, window=5, min_count=1, sg=0)
```

```python
print(model_X)
```

```python
print(tokenized_sentences_X)
```

```python
w1_vec = model_X.wv['ai']
w2_vec = model_X.wv['machine']

print(w1_vec,"\n\n",w2_vec)
```

```python
# calculate cosine similarity between two words
if w1_vec is not None and w2_vec is not None:
  print(cosine_similarity([w1_vec], [w2_vec])[0][0])
else:
  print('None')
```

### **Checking similarity between sentences**

```python
# Function to get the vector for a word
def wordVec(word):
    return model.wv[word] if word in model.wv else None

# Function to get sentence vector by averaging word vectors
def sentVec(sentence):
    vectors = [wordVec(word) for word in word_tokenize(sentence.lower()) if wordVec(word) is not None]
    return np.mean(vectors, axis=0) if vectors else None

# Function to calculate cosine similarity between two sentences
def sentence_similarity(sentence1, sentence2):
    vector1 = sentVec(sentence1)
    vector2 = sentVec(sentence2)
    if vector1 is not None and vector2 is not None:
        return cosine_similarity([vector1], [vector2])[0][0]
    else:
        return None
```

```python
sentence1 = "AI techniques include machine learning and deep learning"
sentence2 = "Word2Vec is a popular technique for word embeddings"
sentence_sim = sentence_similarity(sentence1, sentence2)
print(f"Sentence similarity between:\n'{sentence1}'\nand\n'{sentence2}': {sentence_sim}")
```
