# **Experiment 1**


### **Aim**

Take a corpus from authentic sources like Kaggle and similar sources. Implement the Text Preprocessing (stop word removal, tokenization, stemming, lemmatization, etc.) using SpaCy and NLTK

### **Import libraries**

```python
import kagglehub
import pandas as pd
import os
import re
import string
import spacy

!pip install nltk

import nltk

nltk.download('punkt')

from nltk import sent_tokenize, word_tokenize

nltk.download('stopwords')
from nltk.corpus import stopwords

from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('averaged_perceptron_tagger')
```

### **Loading the data**

```python
# Download latest version
dataset_path = kagglehub.dataset_download("rtatman/blog-authorship-corpus")

# Find the CSV file within the dataset directory
csv_file_path = None  # Initialize csv_file_path to None
for filename in os.listdir(dataset_path):
    if filename.endswith(".csv"):
        csv_file_path = os.path.join(dataset_path, filename)
        break  # Stop searching after finding the first CSV file

# Check if a CSV file was found
if csv_file_path:
    # Read the CSV file
    df_0 = pd.read_csv(csv_file_path)
else:
    print("No CSV file found in the dataset directory.")
```

```python
data = df_0.copy()
data.head()
```

### **Tokenize**

```python
para = data['text'].iloc[0] + " " + data['text'].iloc[1] # forming a sample data from corpus
print(para)
```

**Sentence Tokenize**

```python
sentences = sent_tokenize(para)
print(sentences)
```

**Word Tokenize**

```python
words = word_tokenize(sentences[0])
print(words)
```

### **Toggle case**

**Lower Case**

```python
para_lower = para.lower()
print(para_lower)
```

**Upper Case**

```python
para_upper = para.upper()
print(para_upper)
```

### **Stop word removal**

```python
STOPWORDS = set(stopwords.words('english')) # load english stopwords from nltk library
print(STOPWORDS)
```

```python
words_filtered = []

for w in words:
  if w not in STOPWORDS:
    words_filtered.append(w)
```

```python
print(words_filtered)
```

### **Substituting digit by space**

```python
para_sub = re.sub(r'\d+', ' ', para_lower)
print(para_sub)
```

### **Punctuation removal**

```python
para_new = para.translate(str.maketrans("", "", string.punctuation))
print("Para without punctuations: ",para_new)
```

### **Stemming**

```python
sample_words = ["wait","waiting","waited","waits","listening","history",
                "ability","connect","connected","connections","connecting",
                "connection","sing","caresses","ponies","ties","cats","unicorn",
                "happy","sky","digitizer","sensibiliti"]
```

**Porter Stemmer**

```python
ps = PorterStemmer()
for w in sample_words:
  rootWord = ps.stem(w)
  print(w, " : ", rootWord)
```

**Lancaster stemmer**

```python
ls = LancasterStemmer()
for w in sample_words:
  rootWord = ls.stem(w)
  print(w, " : ", rootWord)
```

**Snowball stemmer**

```python
ss = SnowballStemmer(language='english')
for w in sample_words:
  rootWord = ss.stem(w)
  print(w, " : ", rootWord)
```

### **Lemmatization**

**Using wordnet lemmatizer**

```python
lemmatizer = WordNetLemmatizer()
for w in sample_words:
  print(w, " : ", lemmatizer.lemmatize(w))
```

**Using POS tag**

```python
for w in sample_words:
  print(w, " : ", lemmatizer.lemmatize(w, 'v')) # 'v' = wordnet.VERB
```

### **Observations and Learning**

Stemming is process that cuts off prefixes and suffixes (i.e., affixes) to reduce a word to its root form. But the root form loses the meaning or it is not diminished to a proper English word. Lemmatization makes sure that it does not lose its meaning.

### **Conclusion**

Successfully performed preprocessing techniques like stopword removal, stemming and lemmatization.
