### **Find all unigrams, bigrams and trigrams present in the given corpus**

```python
text = "Had your contract mobile 11 Mnths? Latest Motorola, Nokia etc. all FREE! Double Mins & Text on Orange tariffs. TEXT YES for callback, no to remove from records"
```

**Defining function to generate ngrams**

```python
def generate_ngrams(text, n): # n is words_number, will detect words based on space
    words = text.split()
    ngrams = []
    for i in range(len(words) - n + 1):
        ngrams.append(words[i:i+n])
    return ngrams
```

**Generating unigrams**

```python
generate_ngrams(text, 1)
```

**Generating bigrams**

```python
generate_ngrams(text, 2) # given parameter passed 2 so its bigram
```

**Generating trigrams**

```python
generate_ngrams(text, 3)
```

### **Find the probability using bigrams:**

```python
from collections import defaultdict

def bigram_prob(sentence, training_data):
  # dictionary to store bigram counts
  bigram_counts = defaultdict(lambda: defaultdict(int))

  # counting bigrams in the training data
  for text in training_data:
    words = text.split()
    for i in range(len(words) - 1):
      bigram_counts[words[i]][words[i + 1]] += 1

  probability = 1.0
  words = sentence.split()
  for i in range(len(words) - 1):
    prev_word = words[i]
    current_word = words[i + 1]
    count = bigram_counts[prev_word][current_word]
    total_count = sum(bigram_counts[prev_word].values())
    if total_count > 0:
      probability *= count / total_count
    else:
      probability = 0  # Handle cases where the previous word is not found in the training data. If a bigram is not found, the probability becomes 0

  return probability
```

```python
sentence = "This is my cat"
training_data = [
    'This is a dog',
    'This is a cat',
    'I love my cat',
    'This is my name'
]

probability = bigram_prob(sentence, training_data)
print(f"The probability of the sentence '{sentence}' is: {probability}")
```

### **Predict the next word based on the previous two words(trigrams)**

```python
def predict_next_word(previous_words, training_data):
  trigram_counts = defaultdict(lambda: defaultdict(int))

  # counting trigrams in the training data
  for text in training_data:
    words = text.split()
    for i in range(len(words) - 2):
      trigram_counts[words[i], words[i + 1]][words[i + 2]] += 1

  # predicting the next word based on the previous two words
  next_word = None
  max_count = 0
  for word, count in trigram_counts[tuple(previous_words)].items():
    if count > max_count:
      next_word = word
      max_count = count

  return next_word
```

```python
training_data = [
    "This is a dog",
    "This is a cat",
    "I love my cat",
    "This is my name"
]

previous_words = ["This", "is"]
predicted_word = predict_next_word(previous_words, training_data)
print(f"Predicted next word after '{previous_words[0]} {previous_words[1]}': {predicted_word}")
```

### **Calculate the perplexity of a test text based on n-gram probabilities**

```python
import math

def calculate_perplexity(test_text, n, training_data):
  # Tokenize the test text
  words = nltk.word_tokenize(test_text)

  # Calculating count of n-gram
  ngram_counts = defaultdict(lambda: defaultdict(int))
  for text in training_data:
    tokens = nltk.word_tokenize(text)
    for i in range(len(tokens) - n + 1):
      ngram = tuple(tokens[i:i + n])
      if len(ngram) == n:
        ngram_counts[ngram[:-1]][ngram[-1]] += 1

  # finding probability & perplexity
  total_probability = 1.0
  for i in range(len(words) - n + 1):
    ngram = tuple(words[i:i + n])
    if len(ngram) == n:
      context = ngram[:-1]
      next_word = ngram[-1]
      count = ngram_counts[context][next_word]
      total_count = sum(ngram_counts[context].values())
      if total_count > 0:
        probability = count / total_count
      else:
        probability = 1e-10  # Handle cases where the n-gram is not found
      total_probability *= probability

  perplexity = math.pow(total_probability, -1 / len(words))
  return perplexity
```

```python
test_text = "This is my cat"
training_data = [
    "This is a dog",
    "This is a cat",
    "I love my cat",
    "This is my name"
]
n = 2  # bigram

perplexity = calculate_perplexity(test_text, n, training_data)
print(f"Perplexity of the test text: {perplexity}")
```

### **Observations and Learning**

N-gram can be defined as the contiguous sequence of n items from a given sample of text or speech. The items can be letters, words, or base pairs according to the application. The N-grams typically are collected from a text or speech corpus (A long text dataset). The perplexity value indicates how well the model predicts the test text. Lower perplexity values suggest that the model is better at predicting the test data.
