### **Extracting text from PDF file**

```python
reader = PdfReader('/content/beautifulsoup.pdf')
print(len(reader.pages))
```

```python
# reading first page
pagereader = reader.pages[0]
print(pagereader.extract_text())
```

```python
# read contents of all pages
no_of_pages = len(reader.pages)
print(no_of_pages)
for i in range(len(reader.pages)):
    page = reader.pages[i]
    print(page.extract_text())
```

### **Convert pdf pages to image**

```python
doc = fitz.open('/content/beautifulsoup.pdf')
print(doc.page_count)
```

```python
# convert first page of pdf to image
page = doc.load_page(0)
pix = page.get_pixmap()
pix.save(f"page_{page.number}.png")
```

### **Web Scrapping**

```python
URL = "http://www.youshouldhaveseenthis.com/"
response = requests.get(URL)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,"html.parser")
soup
```

```python
# text data
article =''
for i in soup.findAll('div'):
    article=article+''+i.text
print(article)
```

```python
# extracting all the paragraph texts
paragraphs = soup.find_all("p")
for paragraph in paragraphs:
    print(paragraph.text)
```

```python
# extracting links
links = soup.find_all("a")
for link in links:
    print(link.get("href"))
```
