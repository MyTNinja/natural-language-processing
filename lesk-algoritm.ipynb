### **Aim**

Word Sense Disambiguation (WSD) and its implementation using Lesk Algorithm

### **Lesk Algorithm for Word Sense Disambiguation**

```python
import nltk
from nltk.wsd import lesk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
c = lesk(word_tokenize('face of earth'),'face')
print(c,c.definition())
```

```python
cw = lesk(word_tokenize('face mask'),'face')
print(cw,cw.definition())
```

```python
c2 = lesk(word_tokenize('cricket bat is made of wood'),'bat')
print(c2,c2.definition())
```

```python
c3 = lesk(word_tokenize('bat is a mammal that has wings and can fly'),'bat')
print(c3,c3.definition())
```

```python
c4 = lesk(word_tokenize('earth is flat'),'flat')
print(c4,c4.definition())
```

```python
c5 = lesk(word_tokenize('plane flies through a building'),'plane')
print(c5,c5.definition())
```

```python
c6 = lesk(word_tokenize('a plane, that is an aircraft with 2 engines, flies through a building'),'plane')
print(c6,c6.definition())
```

### **Word Sense Disambiguation (without lesk)**





```python
from nltk.corpus import wordnet as wn

def get_word_senses(word):
    senses = wn.synsets(word)
    return senses

def display_senses(word):
    senses = get_word_senses(word)
    print(f"Word '{word}' has the following senses:")
    for sense in senses:
        print(f"{sense.name()}: {sense.definition()}")

word = "bank"
display_senses(word)
```

### **Observations and Learning**

Word Sense Disambiguation (WSD) is the task of identifying which sense of a word is being used in a given context. This is crucial for understanding the meaning of a sentence correctly when a word has multiple meanings and the intended meaning depends on surrounding words. WSD helps to resolve ambiguity and improve the accuracy of NLP applications. The Lesk Algorithm is a knowledge-based method used for Word Sense Disambiguation. It works by finding the sense of a word that has the most overlap between its dictionary definition and the context in which the word occurs.
