### **Importing Libraries**

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

from sklearn.model_selection import train_test_split

import nltk
nltk.download('punkt')
from nltk.util import ngrams

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import classification_report
```

### **OneHot Encoding**

```python
df = pd.DataFrame({'color': ['red', 'blue', 'green', 'green','red', 'blue']})
df
```

```python
onehot_encoder = OneHotEncoder()
```

```python
onehot_encoded = onehot_encoder.fit_transform(df[['color']]).toarray()

onehot_encoded_df = pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names_out())
onehot_encoded_df
```

```python
pd.concat([df, onehot_encoded_df],axis=1)
```

### **Label Encoding**

```python
label_encoded = label_encoder.fit_transform(df['color'])

label_encoded_df = pd.DataFrame(label_encoded, columns=['label_encoded'])
label_encoded_df
```

**Inference:**

### **Ordinal Encoder**

```python
df = pd.DataFrame({'size': ['small', 'medium', 'large', 'small', 'medium', 'large', 'small', 'medium', 'large']})
df
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
sizes = df[['size']]
encoded_sizes = ordinal_encoder.fit_transform(sizes)

df['encoded_size'] = encoded_sizes
print(df)
```

### **Bag of Words**

```python
v = CountVectorizer(ngram_range=(1,2))

#Build a BOW representation for the corpus
bow_rep = v.fit(["Ram is looking for a job"])

#Look at the vocabulary mapping
print("Our vocabulary: ", v.vocabulary_)

#see the BOW rep for first 2 documents
print("BoW representation for 'Ram is looking for a job': ", bow_rep.transform(["Ram is looking for a job"]).toarray())

#Get the representation using this vocabulary, for a new text
temp = v.transform(["raju is looking for 63"])
print("Bow representation for 'raju is looking for 63':", temp.toarray())
```

**Using on spam dataset**

```python
# loading of 5000 emails from the spam.csv into the dataframe
df = pd.read_csv("/content/spam.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
# 'spam' is marked as 1 and all other values are marked as 0
df['spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.shape
```

```python
df.head()
```

```python
x = df['Message']
y = df.spam # target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

**Converting text into bag of words**

```python
v=CountVectorizer()
x_train_cv=v.fit_transform(x_train.values)
x_train_cv
```

```python
x_train_cv.toarray()
```

```python
v.get_feature_names_out()
```

```python
# retrieve and display 10 of the vocabulary
for i, (key, value) in enumerate(v.vocabulary_.items()):
  if i >= len(v.vocabulary_) - 10:
    print(f"{i}: {key}, {value}")
```

**Using bag of words in Naive Bayes model**

```python
model=MultinomialNB()
model.fit(x_train_cv, y_train)
```

```python
x_test_cv=v.transform(x_test)
```

```python
y_pred=model.predict(x_test_cv)
print(classification_report(y_test,y_pred))
```

### **Bag of N Grams**

```python
text = "Natural language processing is a branch of computer science and linguistics."
```

**Logical method**

```python
def generate_ngrams(text, words_number):
    words = text.split()
    output = []
    for i in range(len(words) - words_number + 1):
        output.append(words[i:i+words_number])
    return output
```

```python
generate_ngrams(text, 3)
```

**Using package**

```python
n_grams = ngrams(sequence=nltk.word_tokenize(text), n=3)
for grams in n_grams:
    print(grams)
```

**Inference:**

Using n-grams allows the model to capture more contextual information than single words alone, improving its ability to understand and represent the text's structure and meaning.

### **Observations and Learning**

Bag of n grams is a feature engineering techniques that converts the textual data into data appropriate for fitting models. It makes groups of set of n words from the vocabulary of the corpus. For Bag of words, n = 1
